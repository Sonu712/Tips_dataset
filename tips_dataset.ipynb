```python
import matplotlib.pyplot as plt
import seaborn as sns
f,[[ax1,ax2],[ax3,ax4]]=plt.subplots(figsize=(10,8),nrows=2,ncols=2)
df=sns.load_dataset("tips")
print(df)
sns.swarmplot(x='total_bill',y='tip',data=df,ax=ax1)
sns.swarmplot(x='total_bill',y='tip',data=df,ax=ax2,hue="sex")
sns.swarmplot(x='total_bill',y='size',data=df,ax=ax3)
sns.swarmplot(x='total_bill',y='size',data=df,ax=ax4,marker='d',palette='Set1')
```
